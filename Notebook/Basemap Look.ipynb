# Initial Look at Basemap
## Create a simple world with coastlines

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map = Basemap()

map.drawcoastlines()

plt.show()
```

## Create a globe with land and oceans colored

```python
map = Basemap(projection='ortho', 
              lat_0=0, lon_0=0)

#Fill the globe with a blue color 
map.drawmapboundary(fill_color='aqua')
#Fill the continents with the land color
map.fillcontinents(color='coral',lake_color='aqua')

map.drawcoastlines()

plt.show()
```

## Try for Chicago

```python
# setup Lambert Conformal basemap.
m = Basemap(llcrnrlon=-90.0378,llcrnrlat=40.6046,urcrnrlon=-85.4277,urcrnrlat=45.1394,
                projection='merc')
# draw coastlines.
m.drawcoastlines()
m.drawstates()

m.drawmapboundary(fill_color='white')

plt.title("City of Chicago")

plt.show()
```
